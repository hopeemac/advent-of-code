```python
def is_valid(pw):
    num_letter = pw[2].count(pw[1])
    if num_letter >= int(pw[0][0]) and num_letter <= int(pw[0][1]):
        print('{} with count of {} is valid'.format(pw, num_letter))
        return True
    else:
        return False
```

## Part 2

```python
def is_valid(pw):
    pw_letter = pw[1]
    pw_str = pw[2]
    pw_index1 = int(pw[0][0]) - 1
    pw_index2 = int(pw[0][1]) - 1
    
    num_matches = 0
    if pw_str[pw_index1] == pw_letter:
        num_matches += 1
        # print('match position 1 num_matches total {}'.format(num_matches))
    if pw_str[pw_index2] == pw_letter:
        num_matches += 1
        # print('match position 2 num_matches total {}'.format(num_matches))
    
    if num_matches == 1:
        print('{} is valid, {}'.format(pw, num_matches))
        return True
    else:
        return False
```

```python
pw = ([1, 3], 'a', 'abcde')
assert is_valid(pw)
```

```python
pw = ([1, 3], 'b', 'cdefg')
assert ~is_valid(pw)
```

```python
pw = ([2, 9], 'c', 'ccccccccc')
assert ~is_valid(pw)
```

```python
num_valid = 0
for pw in input:
    if is_valid(pw):
        num_valid += 1
```
