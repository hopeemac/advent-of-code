## Part 2

```python
f = open("input.txt", 'r')
input = f.read().splitlines() # gets rid of the newline character
```

```python
import math
max_slope1 = 7
field_multiplier = math.ceil((n_rows*max_slope1) / n_columns)
field_multiplier
```

```python
input = [x*field_multiplier for x in input]
```

```python
def calculate_trees_by_slope(input, slope):
    mark = (0,0)
    num_trees = 0
    for step in range(n_rows-1):
        mark = (mark[0] + slope[0], mark[1] + slope[1])
        # print(mark)
        if mark[0] > n_rows:
            # past end of trail
            return num_trees
        else:
            at_mark = input[mark[0]][mark[1]]
            if at_mark == '#':
                num_trees += 1
    return num_trees
```

```python
slope = (2,1)
calculate_trees_by_slope(input, slope)
```

```python
slopes = [(1,1), (1,3), (1,5), (1,7), (2,1)]
trees = []
for s in slopes:
    n_trees = calculate_trees_by_slope(input, s)
    print(s, n_trees)
    trees.append(n_trees)
```
